# Financial Performance (Synthetic)
Generate monthly revenue/cost/profit and visualize KPIs.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
months = pd.date_range('2022-01-01', periods=24, freq='M')
revenue = np.linspace(50000,90000,24) + np.random.normal(0,3000,24)
cost = revenue * (0.62 + np.random.normal(0,0.02,24))
profit = revenue - cost
df = pd.DataFrame({'month':months,'revenue':revenue,'cost':cost,'profit':profit})
plt.plot(df['month'], df['revenue'], label='Revenue')
plt.plot(df['month'], df['profit'], label='Profit')
plt.legend(); plt.xticks(rotation=45); plt.show()
```
